```python
# Importa bibliotecas
import pandas as pd
import numpy as np
import json
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

# Importa warnings
import warnings
warnings.filterwarnings('ignore')
```

# Importação e expansão de bases JSON

```python
# Define configurações do pandas
pd.set_option('display.max_columns', None)
```

```python
# Define caminho
raw_data_path = Path('data/raw')

# Verifica arquivos presentes no diretório
print("Arquivos disponíveis:", list(raw_data_path.iterdir()))
```

```python
# Cria função que carrega um JSON estruturado em dicionário e converte em df
def carregar_e_processar_json(nome_arquivo):
    caminho = Path('data/raw') / nome_arquivo

    with open(caminho, 'r', encoding='utf-8') as f:
        dados = json.load(f)

    # Se o JSON for um dicionário e o valor da primeira chave for uma lista
    if isinstance(dados, dict):
        primeiro_valor = list(dados.values())[0]
        if isinstance(primeiro_valor, list):
            df = pd.DataFrame(primeiro_valor)
        else:
            df = pd.DataFrame.from_dict(dados, orient='index')
            df.reset_index(inplace=True)
            df.rename(columns={'index': 'id'}, inplace=True)
    elif isinstance(dados, list):
        df = pd.DataFrame(dados)
    else:
        raise ValueError("Formato de JSON não suportado.")

    return df
```

```python
# Cria um for loop que itera pelos arquivos .json na pasta,
# carrega cada um como df, normaliza a estrutura, atribui um nome 
# e armazena no dicionário 'dataframes'

dataframes = {}

for arquivo in raw_data_path.glob('*.json'):
    nome_base = arquivo.stem  # ex: 'vagas'
    nome_variavel = f'df_{nome_base}'  # ex: 'df_vagas'
    df = carregar_e_processar_json(arquivo.name)
    dataframes[nome_variavel] = df

# Transforma os itens do dicionário em variáveis
globals().update(dataframes)

# Exibe os nomes das variáveis criadas e o tamanho de cada df
for nome, df in dataframes.items():
    print(f'{nome}: {df.shape}')
```

```python
# Cria funções por df para planificar as bases
def planificar_vagas(df):
    df = df.join(pd.json_normalize(df['informacoes_basicas']))
    df = df.join(pd.json_normalize(df['perfil_vaga']))
    df = df.join(pd.json_normalize(df['beneficios']))
    return df.drop(columns=['informacoes_basicas', 'perfil_vaga', 'beneficios'])

def planificar_prospects(df):
    df = df.explode('prospects', ignore_index=True)
    df = pd.concat([df.drop(columns=['prospects']), pd.json_normalize(df['prospects'])], axis=1)
    return df

def planificar_applicants(df):
    for coluna in ['infos_basicas', 'informacoes_pessoais', 'informacoes_profissionais', 'formacao_e_idiomas']:
        df_norm = pd.json_normalize(df[coluna])
        df_norm.columns = [f"{coluna}_{subcol}" for subcol in df_norm.columns]
        df = df.join(df_norm)
        df.drop(columns=[coluna], inplace=True)
    return df

# Aplica as funções e cria novas variáveis
df_vagas_exp = planificar_vagas(df_vagas)
df_prospects_exp = planificar_prospects(df_prospects)
df_applicants_exp = planificar_applicants(df_applicants)
```

```python
# Define valores positivos e negativos claros para mapeamento da variável target
valores_positivos = ['Contratado pela Decision', 
                     'Aprovado', 
                     'Contratado como Hunting']
valores_negativos = ['Não Aprovado pelo Cliente', 
                     'Não Aprovado pelo RH', 
                     'Não Aprovado pelo Requisitante',
                     'Recusado']  

# Filtra apenas situações com decisão clara
filtro_definido = df_prospects_exp['situacao_candidado'].isin(valores_positivos + valores_negativos)

# Cria DataFrame filtrado
df_filtrado = df_prospects_exp.loc[filtro_definido].copy()

# Cria variável binária target
df_filtrado['target'] = df_filtrado['situacao_candidado'].apply(lambda x: 1 if x in valores_positivos else 0)
```

# Pré-processamento de texto e vetorização

```python
# Importa módulos
import re
import string
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Baixa stopwords em português
nltk.download('stopwords')
stopwords_pt = set(stopwords.words('portuguese'))
```

```python
# Cria uma cópia do df_nlp para testes sem stemming
df_nlp_no_stem = df_nlp.copy()

# Define pontuações a remover
pontuacoes_extras = ['–', '•', '●', '•', '”', '“', '´', '‘', '’', '°']

# Lista de palavras irrelevantes ou estruturais que devem ser removidas
palavras_excluir = {
    'nome', 'resumo', 'qualificações', 'qualificacao',
    'currículo', 'curriculo', 'perfil', 'dados',
    'pessoal', 'telefone', 'email', 'objetivo',
    'documento', 'endereço', 'contato', 'nacionalidade',
    'experiência', 'brasil', 'and', 'atividades', ''
}

# Função sem stemming
def preprocessar_sem_stemming(texto):
    if not isinstance(texto, str):
        return ""

    # Minúsculas
    texto = texto.lower()

    # Remove quebras de linha e tabulação
    texto = re.sub(r'\n|\t', ' ', texto)

    # Remove números
    texto = re.sub(r'\d+', '', texto)

    # Remove pontuação padrão
    texto = texto.translate(str.maketrans('', '', string.punctuation))

    # Remove pontuações extras
    for simbolo in pontuacoes_extras:
        texto = texto.replace(simbolo, '')

    # Tokeniza
    tokens = texto.split()

    # Remove stopwords + palavras irrelevantes
    tokens = [t for t in tokens if t not in stopwords_pt and t not in palavras_excluir]

    return ' '.join(tokens)
    
# Aplica a função ao novo df
df_nlp_no_stem['cv_pt_clean'] = df_nlp_no_stem['cv_pt'].apply(preprocessar_sem_stemming)

# Dropa a coluna original
df_nlp_no_stem.drop(columns=['cv_pt'], inplace=True)
```

```python
# Instancia o vetorizador com limite de features
vectorizer = TfidfVectorizer(max_features=2000)

# Aplica a vetorização sobre a coluna limpa
X = vectorizer.fit_transform(df_nlp_no_stem['cv_pt_clean'])

# Separa variável target
y = df_nlp_no_stem['target']
```

### Sparsidade da matriz TF-IDF

- A matriz TF-IDF construída a partir dos currículos apresentou uma **sparsidade de 0.931**, ou seja, **93,1% dos valores da matriz são zeros**.
- Uma alta sparsidade indica que a maioria dos termos aparece em poucos documentos, o que é esperado em contextos com **vocabulário técnico diverso**.
- Embora seja um cenário típico, a alta sparsidade pode gerar **impacto no desempenho de modelos**, especialmente em algoritmos sensíveis à dimensionalidade.
- Algumas estratégias para lidar com a sparsidade incluem:
  - **Redução de dimensionalidade** (ex.: PCA, TruncatedSVD);
  - **Filtragem de termos pouco frequentes**;
  - **Agrupamento ou embeddings** para representações densas.

### Modelos Baseados em Árvores: Random Forest e XGBoost

Anteriormente, foram testados dois algoritmos de aprendizado de máquina baseados em árvores para classificar currículos como aprovados ou não: Random Forest e XGBoost. Ambos os modelos receberam como entrada vetores reduzidos via TruncatedSVD e adotaram estratégias de balanceamento para mitigar o viés da classe majoritária.

---

#### Modelo 1: Random Forest

- **Configuração:**
  - `n_estimators=200`
  - `class_weight='balanced'`
  - Sem limitação de profundidade
- **Desempenho:**
  - **AUC:** 0.664
  - **F1-score classe 1:** 0.37
  - **Recall classe 1:** 0.27
- **Análise:** O modelo teve forte desempenho na detecção da classe 0, mas deixou a desejar na identificação da classe 1. Isso indica um comportamento conservador, mesmo com o ajuste de pesos.

---

#### Modelo 2: XGBoost

- **Configuração:**
  - `n_estimators=200`
  - `max_depth=6`
  - `learning_rate=0.1`
  - `scale_pos_weight` ajustado à proporção das classes
- **Desempenho:**
  - **AUC:** 0.663
  - **F1-score classe 1:** 0.47
  - **Recall classe 1:** 0.44
- **Análise:** O XGBoost apresentou maior equilíbrio entre as classes, sendo mais eficaz na detecção de candidatos aprovados. Apesar da AUC similar ao Random Forest, a distribuição dos erros é mais favorável.

---

### Conclusões

- O modelo XGBoost mostrou-se mais **sensível à classe positiva**, o que pode ser desejável em problemas com desbalanceamento e foco na **detecção correta de aprovados**.
- A Random Forest, mesmo balanceada, apresentou **tendência à classe negativa**, o que pode ser ajustado com tuning adicional (ex.: `max_depth`, `min_samples_leaf`).
- A seguir, será aplicado o RandomizedSearchCV para otimização de parâmetros.


```python
# Importa módulos
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint

# Define o modelo base
modelo_base = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    random_state=42,
    n_jobs=-1
)

# Estima o scale_pos_weight com base no desbalanceamento da base
scale = len(y_train[y_train == 0]) / len(y_train[y_train == 1])

# Define o espaço de busca de hiperparâmetros
param_distribuicao = {
    'n_estimators': randint(100, 500),                
    'learning_rate': uniform(0.01, 0.2),              
    'max_depth': randint(3, 10),                     
    'min_child_weight': randint(1, 10),               
    'subsample': uniform(0.6, 0.4),                  
    'colsample_bytree': uniform(0.6, 0.4),           
    'gamma': uniform(0, 5),                          
    'scale_pos_weight': [scale]                      # peso da classe minoritária
}

# Configura o RandomizedSearchCV
busca = RandomizedSearchCV(
    estimator=modelo_base,
    param_distributions=param_distribuicao,
    n_iter=30,                         
    scoring='f1',                      # otimiza para F1-score
    cv=5,                              
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Executa a busca
busca.fit(X_train, y_train)

# Melhor modelo encontrado
melhor_modelo = busca.best_estimator_

# Avalia no conjunto de teste
y_pred = melhor_modelo.predict(X_test)
y_proba = melhor_modelo.predict_proba(X_test)[:, 1]

# Exibe métricas
print("Melhores hiperparâmetros:", busca.best_params_)
print(classification_report(y_test, y_pred))
print("Matriz de Confusão:")
print(confusion_matrix(y_test, y_pred))
print(f"AUC: {roc_auc_score(y_test, y_proba):.3f}")
```

### Conclusões do modelo XGBoost otimizado com RandomizedSearchCV

- A otimização de hiperparâmetros com `RandomizedSearchCV` resultou em melhorias importantes no desempenho do XGBoost em relação ao modelo baseline.
- O modelo passou a identificar melhor a classe 1 (aprovado), elevando o **recall para 57%** e o **F1-score para 0.52**, contra 44% de recall e 0.47 de F1-score no modelo não otimizado.
- Apesar de uma leve queda no desempenho da classe 0, o modelo se mostrou **mais sensível e equilibrado**, com uma **AUC de 0.666**, mantendo estabilidade geral.

```python
# Redução de dimensionalidade com SVD
svd = TruncatedSVD(n_components=100, random_state=42)
X_reduced = svd.fit_transform(X)
print(f"Variância explicada total: {svd.explained_variance_ratio_.sum():.2%}")

# Divide os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X_reduced, y, test_size=0.3, stratify=y, random_state=42
)

# Instancia o modelo XGBoost
modelo = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    scale_pos_weight=(len(y_train[y_train == 0]) / len(y_train[y_train == 1])),
    learning_rate=0.1,
    max_depth=6,
    n_estimators=200,
    random_state=42
)

# Treina e avalia o modelo
modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)
y_proba = modelo.predict_proba(X_test)[:, 1]

# Exibe métricas
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(f"AUC: {roc_auc_score(y_test, y_proba):.3f}")
```

```python
# Importa módulos
from sklearn.metrics import precision_score, recall_score, f1_score

# Teste de thresholds 

#-----------------------#

# Intervalo de thresholds para testar
thresholds = np.arange(0.1, 0.91, 0.05)
precisions = []
recalls = []
f1s = []

# Avalia métricas para cada threshold
for t in thresholds:
    y_pred_t = (y_proba >= t).astype(int)
    precisions.append(precision_score(y_test, y_pred_t, pos_label=1))
    recalls.append(recall_score(y_test, y_pred_t, pos_label=1))
    f1s.append(f1_score(y_test, y_pred_t, pos_label=1))

# Plot das métricas por threshold
plt.figure(figsize=(10, 6))
plt.plot(thresholds, precisions, label='Precisão (Classe 1)')
plt.plot(thresholds, recalls, label='Recall (Classe 1)')
plt.plot(thresholds, f1s, label='F1-score (Classe 1)')
plt.axvline(0.5, linestyle='--', color='gray', label='Threshold padrão')
plt.xlabel("Threshold")
plt.ylabel("Valor da métrica")
plt.title("Métricas por Threshold - Classe 1")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Importa módulo
from nltk.stem import RSLPStemmer

# Baixa módulos do nltk
nltk.download('stopwords')
nltk.download('rslp')

# Inicializa stemmer e stopwords
stemmer = RSLPStemmer()
stopwords_pt = set(stopwords.words('portuguese'))

# Pontuações adicionais a remover
pontuacoes_extras = ['–', '•', '●', '”', '“', '´', '‘', '’', '°']

# Palavras irrelevantes a excluir
palavras_excluir = {
    'nome', 'resumo', 'qualificações', 'qualificacao',
    'currículo', 'curriculo', 'perfil', 'dados',
    'pessoal', 'telefone', 'email', 'objetivo',
    'documento', 'endereço', 'contato', 'nacionalidade',
    'experiência', 'brasil', 'and', 'atividades', ''
}

# Função com stemming
def preprocessar_com_stemming(texto):
    if not isinstance(texto, str):
        return ""

    # Minúsculas
    texto = texto.lower()

    # Remove quebras de linha e tabulação
    texto = re.sub(r'\n|\t', ' ', texto)

    # Remove números
    texto = re.sub(r'\d+', '', texto)

    # Remove pontuação padrão
    texto = texto.translate(str.maketrans('', '', string.punctuation))

    # Remove pontuações extras
    for simbolo in pontuacoes_extras:
        texto = texto.replace(simbolo, '')

    # Tokeniza
    tokens = texto.split()

    # Remove stopwords + palavras irrelevantes
    tokens = [t for t in tokens if t not in stopwords_pt and t not in palavras_excluir]

    # Aplica stemming
    tokens = [stemmer.stem(t) for t in tokens]

    return ' '.join(tokens)
```

```python
# Recria a base a partir do merge original
df_nlp_test = pd.merge(
    df_filtrado,
    df_applicants_exp,
    how='inner',
    left_on='codigo',
    right_on='infos_basicas_codigo_profissional'
)

# Aplica a função com stemming nas colunas
for col in [
    'cv_pt',
    'informacoes_profissionais_area_atuacao',
    'informacoes_profissionais_conhecimentos_tecnicos',
    'informacoes_profissionais_nivel_profissional'
]:
    df_nlp_test[col] = df_nlp_test[col].fillna('').apply(preprocessar_com_stemming)

# Cria novamente a coluna unificada
df_nlp_test['texto_completo'] = (
    df_nlp_test['cv_pt'] + ' ' +
    df_nlp_test['informacoes_profissionais_area_atuacao'] + ' ' +
    df_nlp_test['informacoes_profissionais_conhecimentos_tecnicos'] + ' ' +
    df_nlp_test['informacoes_profissionais_nivel_profissional']
)
```

```python
# Vetorização com TF-IDF (bigramas)
vectorizer = TfidfVectorizer(max_features=3000, ngram_range=(1, 2))
X = vectorizer.fit_transform(df_nlp_test['texto_completo'])
y = df_nlp_test['target']

# Redução de dimensionalidade com SVD
svd = TruncatedSVD(n_components=100, random_state=42)
X_reduced = svd.fit_transform(X)
print(f"Variância explicada total: {svd.explained_variance_ratio_.sum():.2%}")

# Divisão em treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X_reduced, y, test_size=0.3, stratify=y, random_state=42
)

# Instancia o modelo base
scale = len(y_train[y_train == 0]) / len(y_train[y_train == 1])
modelo = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    scale_pos_weight=scale,
    random_state=42
)

# Grid de hiperparâmetros
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [4, 6, 8],
    'learning_rate': [0.05, 0.1, 0.2]
}

# Busca com GridSearchCV 
grid_search = GridSearchCV(
    modelo,
    param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1,
    verbose=1
)

# Treina o modelo
grid_search.fit(X_train, y_train)

# Avaliação do modelo final
print("Melhores parâmetros:", grid_search.best_params_)

# Define melhor modelo e fazer previsões
modelo_otimo = grid_search.best_estimator_
y_pred = modelo_otimo.predict(X_test)
y_proba = modelo_otimo.predict_proba(X_test)[:, 1]

# Exibe métricas
print(classification_report(y_test, y_pred))
print("Matriz de Confusão:")
print(confusion_matrix(y_test, y_pred))
print(f"AUC: {roc_auc_score(y_test, y_proba):.3f}")
```

### Próximos passos: Word Embeddings

Apesar do desempenho satisfatório, as limitações do TF-IDF — principalmente a incapacidade de capturar o **significado contextual das palavras** — motivam a adoção de uma abordagem mais robusta com **embeddings vetoriais densos**.

A partir daqui, o projeto passa a explorar o uso de **Sentence Embeddings**, inicialmente com o modelo pré-treinado **`paraphrase-multilingual-MiniLM-L12-v2`** da biblioteca `sentence-transformers`. O objetivo é verificar se o ganho semântico traz melhorias adicionais nas métricas de classificação.



# Teste modelos com embedding
##### Aplicação de embedding: paraphrase-multilingual-MiniLM-L12-v2.

```python
# Importa módulo
from sentence_transformers import SentenceTransformer

# Carrega embeddings
modelo_sbert = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# Gera embeddings para todos os textos
embeddings = modelo_sbert.encode(df_nlp_test['texto_completo'].tolist(), show_progress_bar=True)

# Target
y = df_nlp_test['target'].values

# Divide dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    embeddings, y, test_size=0.3, stratify=y, random_state=42
)

# Calcula peso para balanceamento
scale = len(y_train[y_train == 0]) / len(y_train[y_train == 1])

# Instancia modelo
modelo = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    scale_pos_weight=scale,
    random_state=42
)

# Grid de parâmetros para GridSearchCV
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [4, 6],
    'learning_rate': [0.05, 0.1],
    'subsample': [0.6, 0.8],
    'colsample_bytree': [0.6, 0.8],
    'min_child_weight': [1, 3]
}

# Instancia o gridsearch
grid_search = GridSearchCV(
    modelo, param_grid, cv=5, scoring='f1', n_jobs=-1, verbose=1
)
grid_search.fit(X_train, y_train)

# Exibe melhores parâmetros
print("Melhores parâmetros:", grid_search.best_params_)

modelo_final = grid_search.best_estimator_
y_pred = modelo_final.predict(X_test)
y_proba = modelo_final.predict_proba(X_test)[:, 1]

# Exibe métricas de avaliação
print(classification_report(y_test, y_pred))
print("Matriz de Confusão:")
print(confusion_matrix(y_test, y_pred))
print(f"AUC: {roc_auc_score(y_test, y_proba):.3f}")
```

### Avaliação: Sentence Embeddings com MiniLM + XGBoost Otimizado

Nesta etapa, foi adotada uma nova abordagem de vetorização textual com foco na **representação semântica** dos documentos. Utilizou-se o modelo `paraphrase-multilingual-MiniLM-L12-v2` da biblioteca `sentence-transformers`, capaz de gerar embeddings vetoriais densos e multilingues com alto poder expressivo. Esse modelo permite capturar o **significado contextual das palavras e frases**, superando limitações tradicionais do TF-IDF.

#### Estratégia adotada:

- Geração de embeddings com o MiniLM para todos os textos já unificados em `texto_completo`;
- Divisão em treino e teste com estratificação;
- Treinamento com o modelo **XGBoost**, otimizado via **GridSearchCV**, explorando os seguintes hiperparâmetros:
  - `n_estimators`, `max_depth`, `learning_rate`,
  - `subsample`, `colsample_bytree`, `min_child_weight`;
- Balanceamento das classes aplicado com `scale_pos_weight`.

#### Resultados principais:

- Apesar do ganho semântico oferecido pelos embeddings, os resultados ficaram **abaixo dos obtidos com a combinação TF-IDF + stemização + colunas da vaga**;
- O modelo apresentou **AUC = 0.618**, com **F1-score = 0.47** para a classe positiva;
- Isso indica que, no contexto atual, a abordagem com embeddings densos **não superou os métodos tradicionais**, embora mantenha valor potencial como componente complementar ou em arquiteturas híbridas futuras.


# Teste com incorporação de dados do df de vagas
##### Aplicação de stemming e utilização de modelo XGBoost.

```python
# Merge entre df_filtrado e df_applicants_exp
df_merge = pd.merge(
    df_filtrado,
    df_applicants_exp,
    how='inner',
    left_on='codigo',  # código do candidato
    right_on='id'      # id do candidato em applicants
)

# Renomeia a coluna nivel profissional em df_vagas_exp
df_vagas_exp = df_vagas_exp.rename(columns={'nivel profissional': 'nivel_profissional'})

# Merge com df_vagas_exp
df_merge = pd.merge(
    df_merge,
    df_vagas_exp,
    how='left',
    left_on='id_x',     # id da vaga
    right_on='id'       # id da vaga em df_vagas_exp
)

# Define as colunas textuais
colunas_candidato = [
    'cv_pt',
    'informacoes_profissionais_area_atuacao',
    'informacoes_profissionais_conhecimentos_tecnicos',
    'informacoes_profissionais_nivel_profissional'
]

colunas_vaga = [
    'tipo_contratacao',
    'prioridade_vaga',
    'estado',
    'nivel_profissional',
    'nivel_academico',
    'nivel_ingles',
    'nivel_espanhol',
    'areas_atuacao',
    'principais_atividades',
    'competencia_tecnicas_e_comportamentais'
]

# Preenche valores nulos com string vazia
for col in colunas_candidato + colunas_vaga:
    if col in df_merge.columns:
        df_merge[col] = df_merge[col].fillna('')
    else:
        print(f"Atenção: coluna {col} não encontrada no DataFrame.")

# Concatena todas as colunas textuais em uma única string
df_merge['texto_completo'] = df_merge[
    [col for col in colunas_candidato + colunas_vaga if col in df_merge.columns]
].agg(' '.join, axis=1)

# Cria df final com texto e target
df_nlp_completo = df_merge[['target', 'texto_completo']].copy()

# Exibe primeiras linhas
df_nlp_completo.head()
```

```python
# Aplica pré-processamento
df_nlp_completo['texto_completo'] = df_nlp_completo['texto_completo'].fillna('').apply(preprocessar_com_stemming)
```

```python
# Vetorização TF-IDF com ngramas (1,2)
vectorizer = TfidfVectorizer(max_features=3000, ngram_range=(1, 2))
X = vectorizer.fit_transform(df_nlp_completo['texto_completo'])
y = df_nlp_completo['target']

# Redução de dimensionalidade com TruncatedSVD
svd = TruncatedSVD(n_components=100, random_state=42)
X_reduced = svd.fit_transform(X)
print(f"Variância explicada total: {svd.explained_variance_ratio_.sum():.2%}")

# Divisão em treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X_reduced, y, test_size=0.3, stratify=y, random_state=42
)

# Calcula peso para balanceamento
scale = len(y_train[y_train == 0]) / len(y_train[y_train == 1])

# Instancia o modelo
modelo = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    scale_pos_weight=scale,
    random_state=42
)

# Grid de hiperparâmetros
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [4, 6, 8],
    'learning_rate': [0.05, 0.1, 0.2]
}

# Configura o GridSearchCV
grid_search = GridSearchCV(
    modelo,
    param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1,
    verbose=1
)

# Treino
grid_search.fit(X_train, y_train)

# Exibe resultados
print("Melhores parâmetros:", grid_search.best_params_)

# Avaliação final com melhor estimador
modelo_otimo = grid_search.best_estimator_
y_pred = modelo_otimo.predict(X_test)
y_proba = modelo_otimo.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))
print("Matriz de Confusão:")
print(confusion_matrix(y_test, y_pred))
print(f"AUC: {roc_auc_score(y_test, y_proba):.3f}")
```

### Avaliação: XGBoost com Pré-processamento + Stemização + Dados das Vagas + TF-IDF (Bigrams)

Nesta etapa, foi realizada uma integração completa das fontes de dados:

- Informações textuais do currículo e perfil do candidato (`cv_pt`, `informacoes_profissionais_*`);
- Informações textuais provenientes da vaga (`tipo_contratacao`, `estado`, `areas_atuacao`, entre outras).

As colunas textuais foram unificadas em um campo único (`texto_completo`), passando pelo seguinte pipeline:

#### Estratégia de pré-processamento:
- Aplicação da técnica de **stemização com `RSLPStemmer`** (`nltk`);
- Vetorização com **TF-IDF** (bigramas, `max_features=3000`);
- Redução de dimensionalidade com **`TruncatedSVD`** (100 componentes);
- Divisão treino/teste com estratificação;
- Modelagem com **XGBoost otimizado via GridSearchCV**.

#### Resultados obtidos:
- **Variância explicada** com SVD: `40.63%`
- **Melhores parâmetros encontrados**: `{'learning_rate': 0.05, 'max_depth': 4, 'n_estimators': 100}`
- **Métricas de avaliação:**

  | Métrica              | Valor |
  |----------------------|--------|
  | AUC                  | 0.719  |
  | F1-score (Classe 1)  | 0.55   |
  | Recall (Classe 1)    | 0.56   |
  | Precisão (Classe 1)  | 0.54   |
  | Acurácia geral       | 0.68   |

#### Conclusão:

Este modelo representou **a melhor performance geral até o momento**, superando tanto as versões anteriores com texto limpo quanto aquelas baseadas exclusivamente em embeddings. A combinação de:

- **Stemização linguística (RSLP)**
- **Dados complementares das vagas**
- **XGBoost otimizado**

se mostrou eficaz para capturar nuances importantes nos textos, gerando um modelo mais robusto e confiável para prever contratações com base em perfis textuais.


# Teste final: clusterização + modelo XGBoost otimizado
##### Aplicação de algoritmo KMeans + modelo XGBoost.

```python
# Importações necessárias
from sklearn.cluster import KMeans

# Clustering com KMeans

# Define número de clusters
n_clusters = 4

# Aplica KMeans sobre a matriz reduzida
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(X_reduced)

# Adiciona cluster como nova feature
X_final = np.concatenate([X_reduced, cluster_labels.reshape(-1, 1)], axis=1)

# Divide os dados
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y, test_size=0.3, stratify=y, random_state=42
)

# Calcula balanceamento
scale = len(y_train[y_train == 0]) / len(y_train[y_train == 1])

# Instancia o modelo
modelo = XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    scale_pos_weight=scale,
    random_state=42
)

# Hiperparâmetros (iguais ao modelo final anterior)
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [4, 6, 8],
    'learning_rate': [0.05, 0.1, 0.2]
}

# Otimização com GridSearchCV
grid_search = GridSearchCV(
    modelo, param_grid, cv=5, scoring='f1', n_jobs=-1, verbose=1
)
grid_search.fit(X_train, y_train)

# Avaliação
print("Melhores parâmetros:", grid_search.best_params_)
modelo_otimo = grid_search.best_estimator_

# Predições
y_pred = modelo_otimo.predict(X_test)
y_proba = modelo_otimo.predict_proba(X_test)[:, 1]

# Métricas
print(classification_report(y_test, y_pred))
print("Matriz de Confusão:")
print(confusion_matrix(y_test, y_pred))
print(f"AUC: {roc_auc_score(y_test, y_proba):.3f}")
```

### Avaliação: XGBoost com Clusterização (KMeans) como Feature Adicional

Nesta última etapa de modelagem, foi testada a hipótese de que **a segmentação não supervisionada dos dados textuais** poderia revelar padrões úteis à classificação supervisionada.

#### Estratégia aplicada:

- Aplicação do algoritmo **KMeans (n_clusters = 4)** sobre os vetores textuais já reduzidos via **SVD**;
- Atribuição de cada candidato a um cluster (feature categórica);
- Concatenação desse rótulo ao vetor reduzido como uma nova feature;
- Treinamento com **XGBoost otimizado via GridSearchCV**, com os mesmos hiperparâmetros da melhor versão anterior.

#### Resultados obtidos:

| Métrica              | Valor |
|----------------------|--------|
| AUC                  | **0.719**  |
| F1-score (Classe 1)  | **0.55**   |
| Recall (Classe 1)    | 0.56   |
| Precisão (Classe 1)  | 0.53   |
| Acurácia geral       | 0.67   |

#### Conclusão:

Apesar da adição da clusterização como variável extra, o desempenho **não apresentou ganhos sobre o modelo anterior**. As principais métricas (F1-score e AUC) se mantiveram idênticas à versão baseada apenas em **stemização + TF-IDF + colunas textuais das vagas**.

Além disso, o uso da clusterização introduz **complexidade computacional adicional**, especialmente durante a etapa de ajuste de hiperparâmetros, sem contrapartida em performance.

Por esse motivo, opta-se por manter como versão final o modelo baseado apenas em:
**XGBoost + TF-IDF (bigrams) + Stemização + Dados das Vagas**, que apresenta melhor equilíbrio entre desempenho e simplicidade para o deploy final.

# Criação da pipeline do melhor modelo XGBoost

```python
# Importa módulos
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, TransformerMixin

# Pré-processador customizado
class TextoPreprocessador(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X.apply(preprocessar_com_stemming)

# Pipeline unificada
pipeline_final = Pipeline([
    ('preprocessamento', TextoPreprocessador()),
    ('tfidf', TfidfVectorizer(max_features=3000, ngram_range=(1, 2))),
    ('svd', TruncatedSVD(n_components=100, random_state=42)),
    ('modelo', XGBClassifier(
        objective='binary:logistic',
        eval_metric='logloss',
        use_label_encoder=False,
        scale_pos_weight=scale,
        random_state=42,
        **grid_search.best_params_
    ))
])

# Treina pipeline diretamente
pipeline_final.fit(df_nlp_completo['texto_completo'], y)

# Salva a pipeline toda de uma vez
import joblib
joblib.dump(pipeline_final, 'models/pipeline_modelo_nlp.pkl')
```
